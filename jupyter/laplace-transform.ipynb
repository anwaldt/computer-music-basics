The Laplace transform is a time frequency transform allowing the analysis of LTI systems
in terms of their impulse response behavior. This is relevant when designing and 
investigating filters or other processing units. The transform is defined by the following integral:

$$
F(s) = \int\limits_0^\infty e^{-s t} f(t) dt  
$$

$s$ is the Laplace operator

$$
s = \sigma + j \omega 
$$

which can be decomposed into two terms:

$$
F(s) = \int\limits_0^\infty e^{-\sigma t} \ e^{-j \omega t} f(t) dt  
$$

The term $e^{-\sigma t}$ is referred to as the *decay term*, whereas 
$ e^{-j \omega t}$ will be called the *frequency term*.
  
  
   


```python

def laplace(t,s,w,r):
        
    out = np.zeros(len(t))
    
    for i in range(len(t)):
        val = t[i]
            
        if r ==1:    
            out[i] = np.real(np.exp(-s*val) * np.exp(-1j * w * val))
        else:
            out[i] = np.imag(np.exp(-s*val) * np.exp(-1j * w * val))
            
    return out


plt.figure()

t = np.linspace(0,1,1000)
    
for i in range(10):    
    
    y = laplace(t,i,2 *np.pi *10,1) + 4*i
    plt.plot(t,y,color = [0.15,0.15,0.15])
    
    y = laplace(t,i,2 *np.pi *10,0) + 4*i
    plt.plot(t,y, color = [0.75,0.75,0.75])
    
plt.xlabel('t / s')
plt.ylabel('$\omega$'); 
    
    

plt.figure()

t = np.linspace(0,1,1000)
    
for i in range(10):    
    
    y = laplace(t,5,2 *np.pi *i*10,1) + 4*i
    plt.plot(t,y,color = [0.15,0.15,0.15])
    
    y = laplace(t,5,2 *np.pi *i*10,0) + 4*i
    plt.plot(t,y, color = [0.75,0.75,0.75])
    
   
plt.xlabel('t / s')
plt.ylabel('$\sigma$');     
```
